[Project Euler Problem # 77](https://projecteuler.net/problem=77): Prime Summations

It is possible to write ten as the sum of primes in exactly five different ways:

7+3  
5+5  
5+3+2  
3+3+2+2  
2+2+2+2+2  

What is the first value which can be written as the sum of primes in over five thousand different ways?

```python
import numpy as np
import math
```

```python
max = 10**5
def make_primes(max):
  doublebreak=False
  yield 2
  yield 3
  for i in range(2, max):
    if i % 2 == 0 or i % 3 == 0:
      continue
    for x in range(1,math.ceil(max**(.5)/6), 6):
      if i % (x+1)==0 or i % (x+5)==0:
        doublebreak=True
        break
    if doublebreak:
      doublebreak=False
      continue
    yield i

```

```python
make_primes(max=max)
```

```python
primes=make_primes(max)
```

```python
if 4 in primes:
  print('yes')
```
