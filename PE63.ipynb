```python
'''
Project Euler 
Powerful digit counts

Problem 63
The 5-digit number, 16807=7^5, is also a fifth power. Similarly, the 9-digit number, 134217728=8^9, is a ninth power.

How many n-digit positive integers exist which are also an nth power?'''
```

```python
#first, note that for x<10, x^n could have n digits.  But 10^n has n+1 digits.  And for any x>10, x^n should continue to have >n+1 digits.
#so, just check x<=10
# But, we could potentially have n be quite large, so let's heuristically pick 200
```

```python
count = 0
for x in range(1,11):
  for n in range(1,200):
    if len(str(x**n)) == n:
      count+=1
print(count)
```
